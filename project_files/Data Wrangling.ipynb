```python
def covid_data_extraction(country_code):
    covid_data=pd.read_csv('https://storage.googleapis.com/covid19-open-data/v3/location/'+country_code+'.csv')
    return covid_data
```

```python
all_data=[]
for country_code in data_frame['company_location'].unique():
    print(country_code)
    try:
        data=covid_data_extraction(country_code)
        all_data.append(data)
        print('Success')
    except:
        print('Not found')
    print('-------------------------------------')
```

```python
final_covid_data=pd.concat(all_data)
```

```python
final_covid_data.reset_index(inplace=True)
```

```python
final_covid_data.drop(['index'],axis=1,inplace=True)
```

```python
pd.set_option('display.max_columns', None)
```

```python
cov19_features=['country_code','date','new_confirmed','new_deceased','new_tested','cumulative_confirmed','cumulative_deceased','cumulative_tested']
```

```python
final_covid_data=final_covid_data[cov19_features]
```

```python
final_covid_data[final_covid_data['country_code']=='US']['cumulative_confirmed'].mean()
```

```python
final_covid_data['year'] = pd.DatetimeIndex(final_covid_data['date']).year
```

```python
final_covid_data.groupby(['year','new_confirmed'],axis=1).head()
```

```python
country_codes=[]
years=[]
means=[]
for country in final_covid_data.country_code.unique():
    for year in [2020,2021,2022]:
        mean=final_covid_data[(final_covid_data['country_code']==country)&(final_covid_data['year']==year)]['new_confirmed'].mean()
        mean=round(mean)
        country_codes.append(country)
        years.append(year)
        means.append(mean)
```

```python
new_df=pd.DataFrame({'country_code':country_codes,'year':years,'mean':means})
```
